```python
def sampleMVN(p_mu,p_v):
    return  np.random.multivariate_normal(mean=p_mu, cov=p_v, check_valid='warn', tol=1e-8,size=1)

def simulateBetsWLearning(princLearn,singleTraj=False):
        
    global m0;   global mD;   global v0;   global vD;   global N_E_EP #SAMPLE COUNTER;
    
    tt=trials #Set no. trials
    if singleTraj:
        c=1
        tt=[1]
    
    final=np.ones((trials,3)); 
    time_prefs=np.ones((trials,STEPS,3));
    
    resources = np.ones((trials,STEPS))

    for tr in range(trials):
        means   =np.zeros((3,STEPS));
        var_s   =np.zeros((STEPS,3,3)); 
        samples =np.zeros((STEPS,3));
        pens=np.zeros(STEPS-1)
        resetAgent(1)

        for i in range(CYCLES):

            for t in range(1,STEPS):
                
                tau=t
                
                #Sampling distribution
                samples = sampleMVN(mD,vD); #samples[t]=sample; #Record samples
#               #Updating resources
                payout= sp.stats.gmean([checkAgentMVN(s) for s in samples]) #Computing payout
                pens[t-1]=payout
                resources[tr][t]=resources[tr][t-1]*payout #Appending to resources

                #Updating the Agents
                mD,vD1,N_E_EP=mD,vD,N_E_EP
#                 sample=np.mean(samples,axis=0)
#                 mD,vD1,N_E_EP=learnTrivGaussian(np.copy(m0),N_E_EP,np.copy(v0),\
#                                                 sample,t,N[t],beta0,'ag',k)
                
                #Updating means and variances
                var_s[t]=np.copy(vD);
            
            #RECORDS THE FINAL VALUE OF EACH TRIAL
            final[tr]=mD
            time_prefs[tr]=np.transpose(means);        

    #For running single trajectories
    return resources
    return histData(final,time_prefs)

#     #For getting spectra of agents
#     return final_ps,final_as

def learnTrivGaussian(p0,w,v,s,t,N,b0,ag,kk):
    sample=[s[2],s[1]] #Shows a to p and p to a
    #Adds sample to total data history
    w[t]=sample
    print(t)
    print(w[:t],n2D[:t+1])
    wBar=np.divide(np.cumsum(w[:t],axis=0),n2D[:t])

    #Computes cumulative mean data for agent based on data weighted relative to prior,
    # and relative to one another (NOT YET)
    m = np.divide(np.add(b0*kk*p0,np.multiply(n2D[t],wBar[t])),(b0*kk+N))

    
    return m,v,w


def rvToList(rv,x):
    if str(type(rv))=="<class 'list'>":
        return np.array([rv_i.pmf(x) for rv_i in rv])
    else:
        return rv.pmf(x)

def resetAgent(p):
    global N_E_EP
    N_E_EP=np.ones((STEPS,2))
    global mD
    mD = m0
    global vD
    vD = v0
  
    
def simulateDynamics(N,order,corr,W,p_success,typ,switch_times):
    resetNs(order,'p');resetNs(order,'a')
    resources=simulateBetsWLearning()
    return resources,agents,[extractPMF(s,order) for s in sgr],signal_probs,Gs

#REWARDS FUNCTION IS NOT NORMALIZED PROPERLY
def checkAgentMVN(sample):
    global mult
    reward = stats.norm.pdf(sample[2], loc=pm+v0x[1,2]/v0x[2,2]*(sample[1]-pm),\
                            scale=np.sqrt(v0x[1,1]-v0x[1,2]**2/v0x[2,2]))
    return mult*reward


#Agent 
N_E_EP=np.ones((STEPS,2))

N_COV=[]
#Principal
N_S=[]
N_R=[]
N_SR=[]
```

```python
################
################
####Code generating learning time curves
################
################

def compPrAg(t,k,pk,x0,y0):
    ag=None
    pr=None
    if k!=pk:
        term2=np.log((pk*(k+t))/(k*(pk+t)))*k*pk*(x0-y0)/(k-pk)**2
        ag=( k*x0)/( k+t)+term2+(k*x0-pk*y0)/(k-pk)*t/( k+t)
        pr=(pk*y0)/(pk+t)+term2+(k*x0-pk*y0)/(k-pk)*t/(pk+t)
        
        LC=np.log(pk/k)
        termp=y0*(pk/k-LC-1)
        terma=x0*(k/pk+LC-1)
        den=(k-pk)**2/(k*pk)
#         plt.axhline((terma+termp)/den,alpha=.1,color='k')
    else:
        ag=(2*k**2*x0+(2*k*t+t**2)*(x0+y0))/(2*(k+t)**2)
        pr=(2*k**2*y0+(2*k*t+t**2)*(x0+y0))/(2*(k+t)**2)
    return pr,ag

def compPrAgHongler(t,k,pk,x0,y0):
    ag=None
    pr=None
    if k!=pk:
        term1=(x0+y0)/2
        term2=(x0-y0)
        term3=1/(k-p_k)**2*np.log((p_k*t+k*p_k)/(k*t+k*p_k))\
                +(k+p_k)/(k*p_k*(k-p_k))
        ag=term1+term2*(term3-1/((k-p_k)*(t+k)))
        pr=term1+term2*(term3-1/((k-p_k)*(t+p_k)))
    
    else:
        ag=(2*k**2*x0+(2*k*t+t**2)*(x0+y0))/(2*(k+t)**2)
        pr=(2*k**2*y0+(2*k*t+t**2)*(x0+y0))/(2*(k+t)**2)
    return pr,ag

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(5,3))

t=np.linspace(1,100,100)
pk=5
color=[(i-.15,0,i) for i in [.3,.4,.5,.6,.7]]
x0=1;y0=5
for k,c in zip([1,3.5,10,35,250],color):

#         plt.axhline((x0+y0)/2,alpha=.1,color='k')
    pr,ag = compPrAg(t,k,pk,x0,y0)
    ax1.set_xlim(left=.5,right=4.6)
    ax1.plot(np.log(t),ag,color='k',linestyle='dashed');
    ax1.plot(np.log(t),pr,color=c,linewidth=2.5,label=k)
    ax1.set_ylim(top=5.5,bottom=1)

ax1.set_ylabel("Preference")
ax1.set_xlabel(r"$\log(t)$")

k=5
color=[(.9,i-.1,.2-i/4) for i in np.flip([.1,.25,.4,.55,.7])]
for pk,c in zip(([1,3.5,10,35,250]),color):
    ag=None
    pr=None
    pr,ag = compPrAg(t,k,pk,x0,y0)
    ax2.set_xlim(left=.5,right=4.6)
#     ax2.plot(np.log(t),ag,color='k',linestyle='dashed');
    ax2.plot(np.log(t),pr,color='k',linestyle='dashed')
    ax2.plot(np.log(t),ag,color=c,linewidth=2.5,label=pk);

# ax2.legend(loc="lower center",ncol=5,framealpha=1,title=r'$\alpha$',prop={'size': 6})
# ax1.legend(loc="lower center",ncol=5,framealpha=1,title=r'$\beta$',prop={'size': 6})
ax2.set_ylim(top=5.2,bottom=.8)
ax1.set_ylim(top=5.2,bottom=.8)
plt.tight_layout()
# plt.savefig("dp_param.pdf")
plt.savefig("deterministicparam.pdf")
plt.show()
```
